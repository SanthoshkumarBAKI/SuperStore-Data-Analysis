# Vizualizaion

```python
sns.scatterplot(x=df["Sales"],y=df["Profit"])
plt.title('Profit vs Sales')
plt.show()
```

Sales values soar up to 20,000, while profit values fluctuate between -4,000 and 8,000.

```python
sns.lineplot(x=df['Discount'], y=df['Profit'])
plt.title('Profit by Discount')
plt.show()
```

lineplot represents profit as a function of discount, with a shaded area around the line indicating some measure of variability or confidence interval. The line starts at a profit of around 100 at a discount of 0.0, then fluctuates and generally trends downward, reaching a minimum profit of around -400 at a discount of 0.5, before rising again slightly towards a discount of 0.8.

```python
sns.barplot(x=df["Sales"],y=df["Quantity"])
plt.title("Sales vs Quantity")
```

The x-axis, you have Quantity ranging from 0 to 14.
The y-axis is labeled Sales, ranging from 0 to 20,000. The plot is dotted with numerous blue points, each representing a data set showing the relationship between the quantity of items and their corresponding sales figures.


```python
sns.barplot(x=df["Quantity"],y=df["Sales"])
plt.title("Quantity vs Sales")
plt.show
```

 I see a vivid tapestry of sales distributions, with each box revealing the intricate story of median values, interquartile ranges, and outliers. The whiskers stretch out like arms, embracing the data's natural variation, while outliers stand alone, whispering secrets of unusual sales patterns.

```python
region_sales = df.groupby('Region').sum()['Sales']
region_sales.plot(kind='bar', color='skyblue')
plt.title('Sales by Region')
plt.ylabel('Sales')
plt.show()
```

X-Axis (Sales): Increments of 10,000, ranging from 0 to 70,000.
Y-Axis (Regions): Lists the regions Central, East, South, and West.
Each bar represents the sales for a specific region, with the following approximate values:
- *Central*: 40,000
- *East*: 60,000
- *South*: 30,000
- *West*: 70,000


```python
region_sales = df.groupby('Sales').sum()['Profit']
region_sales.plot(kind='box', color='green')
plt.title('Profit by Sales')
plt.ylabel('Profit')
plt.show()
```


The plot shows the distribution of Profit grouped by Sales. Most profits are near the median, with some extreme outliers indicating unusually high or low profits. The box represents the middle 50% of profits, and outliers are visible beyond the whiskers

```python
category_profit = df.groupby('Category').sum()['Profit']
category_profit.plot(kind='bar', color='orange')
plt.title('Profit by Category')
plt.ylabel('Profit')
plt.show()
```

X-Axis (Profit): Increments of 20,000, ranging from 0 to 140,000.
Y-Axis (Category): Lists three categories: Furniture, Office Supplies, and Technology.


```python
segment_profit=df.groupby('Segment').sum()['Profit']
segment_profit.plot(kind='bar', color='orange')
plt.title('Profit by Segment')
plt.ylabel('Profit')
plt.show()
```

X-Axis (Profit): Increments of 20,000, ranging from 0 to 140,000.   
Y-Axis (Segments): Lists three segments: Consumer, Corporate, and Home Office.   
Each segment is represented by a blue bar. The chart indicates the following approximate profit values:   
-Consumer: 130,000
-Corporate: 70,000
-Home Office: 40,000

```python
#shipmode vs profit
ship_mode_profit = df.groupby('Ship Mode').sum()['Profit']
ship_mode_profit.plot(kind='bar', color='purple')
plt.title('Profit by Ship Mode')
plt.ylabel('Profit')
plt.show()     
```

X-Axis (Profit): Increments of 2,000, ranging from 0 to 16,000.  
Y-Axis (Shipping Modes): Lists four shipping modes: First Class, Same Day, Second Class, and Standard Class.    
- First Class: 8,000
- Same Day: 2,000
- Second Class: 10,000
- Standard Class: 15,000
